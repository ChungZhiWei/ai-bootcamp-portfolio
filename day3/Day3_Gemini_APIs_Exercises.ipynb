```python
# Exercise 2 — Setup (run this cell first)
from google import genai
from google.genai import types
from dotenv import load_dotenv
load_dotenv()
client = genai.Client()
```

```python
# Task 2.1 — Complete the base chat() function

def chat(prompt, system, temperature=0.7, max_tokens=200):
    """Ask the AI. Returns only the answer string."""
    response = client.models.generate_content(
        model    = 'gemini-3.1-flash-lite',
        contents = prompt,                          # the user's prompt goes here
        config   = types.GenerateContentConfig(
            system_instruction = system,             # the system parameter goes here
            temperature        = temperature,
            max_output_tokens  = max_tokens
        )
    )
    return response.text                           # what attribute holds the text?

# Quick test — should print something like: "2 + 2 equals 4."
print(chat('What is 2+2?', 'You are a helpful assistant.'))
```

### Task 3.2 — Build a Simple Memory Function

**Your Task:** Fill in the blanks to make the function maintain conversation history.

The three steps, in order:
1. Add the user's question to `history` **before** the API call
2. Pass `history` as the `contents` argument (the system prompt stays **separate**, passed via `system_instruction`)
3. Add the AI's reply to `history` **after** the API call — using role `'model'`, not `'assistant'`

> Replace `[YOUR NAME]` and `[YOUR MAJOR]` with your real details!

<div style="background: #fff8e1; border: 1px solid #ffe082; border-radius: 8px; padding: 10px 16px; font-size: 13px;">
 <strong> Note:</strong> Gemini keeps the system prompt <em>outside</em> the history list, and uses <code>'user'</code> / <code>'model'</code> roles instead of OpenAI's <code>'user'</code> / <code>'assistant'</code>. Each history entry looks like <code>{'role': 'user', 'parts': [{'text': '...'}]}</code>.
</div>

```python
# Task 3.2 — Build a memory function

SYSTEM_PROMPT = 'You are a friendly study assistant for DigiPen students.'
history = []   # starts empty — the system prompt is NOT stored in here

def chat_with_memory(question):
    """Add question to history, call API, add answer to history, return answer."""

    # Step 1: Add user question to history BEFORE the API call
    history.append({'role': 'user', 'parts': [{'text': question}]})     # ← what method? what goes in 'text'?

    # Step 2: Call API with the FULL history + system prompt
    response = client.models.generate_content(
        model    = 'gemini-3.1-flash-lite',
        contents = history,                                         # ← pass the full history list here
        config   = types.GenerateContentConfig(
            system_instruction = SYSTEM_PROMPT,
            max_output_tokens  = 150
        )
    )
    answer = response.text

    # Step 3: Add AI's reply to history AFTER getting the response (role='model', not 'assistant')
    history.append({'role': 'model', 'parts': [{'text': answer}]})      # ← same method, but which role?

    total_tok = response.usage_metadata.total_token_count
    print(f'  [Turn {len(history)//2} | Messages in history: {len(history)} | Tokens this call: {total_tok}]')
    return answer


# Test it — replace [YOUR NAME] and [YOUR MAJOR] before running!
turns = [
    'My name is zhi wei and I am studying game design.',
    'What should I focus on in my degree?',
    'What was my name again?',
    'How can I use my degree in the Singapore tech industry?'
]

for q in turns:
    print(f'You: {q}')
    print(f'Bot: {chat_with_memory(q)}')
    print()
```

**Reflection — double-click to edit:**

Did the bot remember your name in Turn 3? Yes.

What do you notice about the token count each turn? Why does it grow?

The token count increases each turn. It grows because we are charged tokens for sending and getting a response.

What problem might arise if a conversation went on for 1,000 turns?

The chatbot would stop working as we hit token limit.

### Task 3.3 — The Full `MemoryBot` Class

**Your Task:** No blanks here — run the cell as-is and test it. Notice that `tutor` and `coach` have **completely separate** and independent histories.

```python
# Task 3.3 — MemoryBot class (no blanks — run and observe)

class MemoryBot:
    def __init__(self, personality='helpful'):
        prompts = {
            'helpful': 'You are a friendly, helpful assistant.',
            'tutor':   'You are a patient CS tutor. Use simple examples. Always check understanding.',
            'coach':   'You are an energetic life coach. Always motivate. Start every reply with an emoji.',
        }
        self.system    = prompts.get(personality, prompts['helpful'])
        self.history   = []
        self.turn      = 0
        self.total_tok = 0

    def ask(self, question):
        self.turn += 1
        self.history.append({'role': 'user', 'parts': [{'text': question}]})
        response = client.models.generate_content(
            model    = 'gemini-3.1-flash-lite',
            contents = self.history,
            config   = types.GenerateContentConfig(
                system_instruction = self.system,
                max_output_tokens  = 200
            )
        )
        answer = response.text
        self.total_tok += response.usage_metadata.total_token_count
        self.history.append({'role': 'model', 'parts': [{'text': answer}]})
        return answer

    def status(self):
        # gemini-2.5-flash: ~$0.30/1M input, ~$2.50/1M output tokens — blended estimate below
        cost = self.total_tok * 0.0000006
        return f'Turn {self.turn} | {len(self.history)} messages | {self.total_tok} tokens | ${cost:.6f}'


# Test tutor personality
print('=== TUTOR BOT ===')
tutor = MemoryBot(personality='tutor')
print(tutor.ask('Explain what a variable is.'))
print(tutor.status())
print(tutor.ask('Can you give me a quiz question about variables?'))
print(tutor.status())
print()

# Test coach personality
print('=== COACH BOT ===')
coach = MemoryBot(personality='coach')
print(coach.ask('I feel like giving up on programming. It is too hard.'))
print(coach.status())
```

```python
# Task 4.2 — Build your application
# Replace all ??? with your actual code


# ── Your System Prompt ─────────────────────────────────────────────────────────
MY_SYSTEM_PROMPT = '''
You are a coding specialist named "Tom". 
You are able to explain any code given to you line by line for beginners.
You will always ask them if you want to make any changes to the code after explaining the code.
If user wants to make changes to the code you give you step by step instructions on how to do the changes.
You always start the reply with "I see".
You always reply to finish task with "That is all folks."
'''

# ── Your Bot (function or class) ───────────────────────────────────────────────
#
# No memory needed? Use a simple function:
#
# def my_app(question):
#     response = client.models.generate_content(
#         model    = 'gemini-2.5-flash',
#         contents = question,
#         config   = types.GenerateContentConfig(
#             system_instruction = MY_SYSTEM_PROMPT,
#             temperature        = ???,
#             max_output_tokens  = ???
#         )
#     )
#     return response.text
#
# Need memory? Use a class (see MemoryBot from Task 3.3 as your template):
#
# class MyApp:
#     def __init__(self): ...
#     def ask(self, question): ...

   # Write your implementation here
class MemoryBot:
    def __init__(self):
        self.system    = MY_SYSTEM_PROMPT
        self.history   = []
        self.turn      = 0
        self.total_tok = 0

    def ask(self, question):
        self.turn += 1
        self.history.append({'role': 'user', 'parts': [{'text': question}]})
        response = client.models.generate_content(
            model    = 'gemini-3.1-flash-lite',
            contents = self.history,
            config   = types.GenerateContentConfig(
                system_instruction = self.system,
                max_output_tokens  = 200
            )
        )
        answer = response.text
        self.total_tok += response.usage_metadata.total_token_count
        self.history.append({'role': 'model', 'parts': [{'text': answer}]})
        return answer

    def status(self):
        # gemini-2.5-flash: ~$0.30/1M input, ~$2.50/1M output tokens — blended estimate below
        cost = self.total_tok * 0.0000006
        return f'Turn {self.turn} | {len(self.history)} messages | {self.total_tok} tokens | ${cost:.6f}'


bot = MemoryBot()


# ── 5 Test Questions ───────────────────────────────────────────────────────────
test_questions = [
    '''
    os.environ["OPENAI_API_KEY"] = "your-openai-api-key"
    m = Memory()
    messages = [
        {"role": "user", "content": "Hi, I'm Alex. I love basketball and I'm allergic to peanuts."},
        {"role": "assistant", "content": "Hello Alex! I will keep that in mind."}
    ]
    m.add(messages, user_id="alex")

    search_results = m.search("What sports do I like?", filters={"user_id": "alex"})

    all_memories = m.get_all(user_id="alex")
    ''',
    'The code tells me API key is invalid. HOw do I fix this?',
    'The code say my name is Alex but my name is Bob. How do I fix this?',
    'Can you explain the code with all these fixes?',
    'Thanks, these fixes worked.',
]

for q in test_questions:
    print(f'Q: {q}')
    answer = bot.ask(q)
    print(f'A: {answer}')
    print()

print(bot.status())
# ── Token & Cost Summary ───────────────────────────────────────────────────────
# print(f'Total tokens used: {???}')
# print(f'Estimated cost:    ${??? * 0.0000006:.6f}')
```
